```python
from sklearn import datasets

# List of datasets available in sklearn
dataset_functions = [func for func in dir(datasets) if func.startswith('load_') or func.startswith('fetch_')]

print("Datasets available in sklearn:")
for dataset in dataset_functions:
    print(dataset)

# Load the iris dataset and view the first few rows
iris = datasets.load_iris(as_frame=True).frame
print("\nLoaded 'iris' dataset:")
print(iris.head())
iris.head()
```

# tensorflow

```python
# Tensorflow Datasets
import tensorflow_datasets as tfds
import pandas as pd

# List the datasets available in tensorflow_datasets
datasets_list = tfds.list_builders()

print("Datasets available in tensorflow_datasets:")
for dataset in datasets_list:
    print(dataset)

# Load the 'mnist' dataset
ds, info = tfds.load('mnist', with_info=True, as_supervised=True)

# Convert the dataset to a pandas DataFrame (for the first 5 examples in the training split)
df = tfds.as_dataframe(ds['train'].take(5), info)

print("\nLoaded 'mnist' dataset:")
print(df.head())
```

```python
import pandas as pd
import requests
import pyreadstat

# URL of the SPSS file
url = "https://stats.oarc.ucla.edu/stat/data/binary.sav"

# Download the file
response = requests.get(url)

# Save the file locally
with open("binary.sav", "wb") as f:
    f.write(response.content)

# Load the SPSS file
df, meta = pyreadstat.read_sav("binary.sav")

# Display the DataFrame
df

# save it as csv format
df.to_csv("binary.csv", index=False)
df.tail()
```


# ZIP files

```python
# First Step
!pip install requests
import requests
import pandas as pd
from zipfile import ZipFile
import sqlite3

# Download and unzip the Jobs data
zip_url = 'https://www.stata-press.com/data/aecss/aecss.zip'
response = requests.get(zip_url)

# Save the content of the response to a file
with open('aecss.zip', 'wb') as f:
    f.write(response.content)

file_names = list()
with ZipFile('aecss.zip', mode='r') as zip:
    zip.extractall()
    for file in zip.infolist():
        file_names.append(file.filename)
        print(file.filename, file.compress_size, file.file_size)
```
